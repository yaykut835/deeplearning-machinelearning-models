```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("heart.csv")
```

```python
df.head(3)
```

```python
df=df.drop(columns=["Patient ID", "Hemisphere", "Continent"])
```

```python
df.head(3)
```

```python
df[["sis", "dis"]]= df["Blood Pressure"].str.split("/", expand=True)
df["sis"]= df["sis"].astype(int)
df["dis"]= df["dis"].astype(int)
df=df.drop(columns=["Blood Pressure"])
```

```python
df.head(3)
```

```python
le=LabelEncoder()
df["Diet"]= le.fit_transform(df["Diet"])
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df, columns=["Sex"], drop_first=True)
df=df.drop(columns=["Country"])
df.head(3)
```

```python
df["Sex_Male"]= df["Sex_Male"].astype(int)
df.head(3)
```

```python
y=df["Heart Attack Risk"]
x=df.drop("Heart Attack Risk", axis=1)
x_train, x_test, y_train, y_test=train_test_split(x,y, train_size=0.90, random_state=78)
```

```python
model=RandomForestClassifier(n_estimators=100, random_state=78)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
